# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_school = len(school_data["School ID"].unique())
total_student = school_data_complete["student_name"].count()
total_budget = school_data["budget"].sum()
avg_math = school_data_complete["math_score"].mean()
avg_reading = school_data_complete["reading_score"].mean()
overal_avg_score = (avg_math + avg_reading)/2

#avg_score = (student_data["math_score"]+student_data["reading_score"])/2
#student_data["Average Score"] = avg_score

passing_math_stu = school_data_complete[school_data_complete["math_score"] >= 70]

percent_passing_math = passing_math_stu["student_name"].count()/total_student*100

passing_reading_stu = school_data_complete[school_data_complete["reading_score"] >= 70]

percent_passing_reading = passing_reading_stu["student_name"].count()/total_student*100


passing_overal = (percent_passing_math+percent_passing_reading)/2
        

#percent_passing = passing_overal["student_name"].count()/total_student

district_summary = pd.DataFrame({"Total Schools": [total_school], "Total Students":[total_student],
                                "Total Budget": [total_budget],"Average Math Score": [avg_math],
                                "Average Reading Score":[avg_reading],
                                "%Passing Math":[percent_passing_math],
                                "%Passing Reading":[percent_passing_reading],
                                "%Overal Passing": [overal_avg_score]})
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("{:,}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["%Passing Math"] = district_summary["%Passing Math"].map("{:.2f}%".format)
district_summary["%Passing Reading"] = district_summary["%Passing Reading"].map("{:.2f}%".format)
district_summary["%Overal Passing"] = district_summary["%Overal Passing"].map("{:.2f}%".format)
district_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_group = school_data_complete.groupby("school_name")
total_student = school_group.size()
school_type = school_group["type"].first()
school_type
total_budget = school_group["budget"].first() #.map("{:,}".format)
total_budget
budget_per_student = total_budget/total_student
budget_per_student
average_math_score = school_group["math_score"].sum()/total_student
average_math_score
average_reading_score = school_group["reading_score"].sum()/total_student
average_reading_score
percent_passing_math = passing_math_stu.groupby("school_name").size()/total_student*100
percent_passing_reading = passing_reading_stu.groupby("school_name").size()/total_student*100
percent_passing_reading
overal_passing_rate = (percent_passing_math+percent_passing_reading)/2
school_group_summary = pd.DataFrame({"School Type": school_type, "Total Budget":total_budget,
                                     "Total Students": total_student,
                                   "Budget per Student": budget_per_student,
                                   "Average Math Score": average_math_score,
                                   "Average Reading Score": average_reading_score,
                                   "% Passing Math":percent_passing_math,
                                   "% Passing Reading": percent_passing_reading,
                                   "% Overall Passing Rate": overal_passing_rate})
school_group_summary_d = school_group_summary.copy()
school_group_summary_d["Total Students"] = school_group_summary["Total Students"].map("{:,}".format)
school_group_summary_d["Total Budget"] = school_group_summary["Total Budget"].map("${:,}".format)
school_group_summary_d["Budget per Student"] = school_group_summary["Budget per Student"].map("${:,}".format)
school_group_summary_d["Average Math Score"] = school_group_summary["Average Math Score"].map("${:.2f}".format)
school_group_summary_d["Average Reading Score"] = school_group_summary["Average Reading Score"].map("{:.2f}".format)
school_group_summary_d["% Passing Math"] = school_group_summary["% Passing Math"].map("{:.2f}%".format)
school_group_summary_d["% Passing Reading"] = school_group_summary["% Passing Reading"].map("{:.2f}%".format)
school_group_summary_d["% Overall Passing Rate"] = school_group_summary["% Overall Passing Rate"].map("{:.2f}%".format)
school_group_summary

top_school = school_group_summary_d.sort_values(by= "% Overall Passing Rate", ascending = False)
top_school.head(5)
```

```python
worse_school = school_group_summary_d.sort_values(by= "% Overall Passing Rate", ascending = True)
worse_school.head(5)
```

## Scores by School Spending
